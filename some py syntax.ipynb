```
#some test lists
l1 = ['a', 'b', 'c']
l2 = ['1', '2', '3']
l3 = ['1', '2', '3']

#some test generators
twos = (i for i in range(100) if i%2==0)
fives = (i for i in range(100) if i%5==0)
```

```
#####
# ZIP
#####
```

```
#iterate to print
for i in zip(l1,l2):
  print(i)
```

```
#print zip as list
print(list(zip(l1, l2)))
```

```
#print tuples
for a,b in zip(l1,l2): 
  print(a,b)
```

```
#print as dict
print(dict(zip(l1,l2)))
```

```
#print formatting
for a,b in zip(l2,l3): 
  c = int(a)+int(b)
  print('a: {} + b: {} = c: {}'.format(a,b,c))
```

```
############
# Generator
############
```

```
#print fives generator object, useless
print(fives)
```

```
#iterate and print elements of generator object
[print(i) for i in fives]
[print(i) for i in twos]
```
