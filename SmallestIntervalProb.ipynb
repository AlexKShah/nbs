# Finding The Smallest Interval Of K Sorted Lists
https://www.dailycodingproblem.com/blog/smallest-interval-of-k-sorted-lists/

This interesting and difficult problem was asked by Google recently.

> Given K sorted lists of integers, return the smallest interval (inclusive) that contains at least one element from each list. If there are multiple intervals of the same size, return the one that starts at the smallest number.

For example, given:

[ 
  [0, 1, 4, 17, 20, 25, 31],

  [5, 6, 10],

  [0, 3, 7, 8, 12] 
  ]

The smallest range here is [3, 5], since it contains 4 from the first list, 5 from the second list, and 3 from the third list.



 Input:

List of sorted lists of integers

---
Output:

return integer interval as [start,end]

---
conditionals:

smallest inclusive interval

contains at least one element from each list

if multiple intervals then choose min start

```
K = [[0, 1, 4, 17, 20, 25, 31], [5, 6, 10], [0, 3, 7, 8, 12]]
```

```
# start and end to find the smallest and largest starting values
start_val = K[0][0]
end_val = K[0][0]

for n in K:
  if n[0] > end_val:
    end_val = n[0] 
  if n[0] < start_val:
    start_val = n[0]
    
while start_val < end_val:
  for n in K:
    for i in n:
      if start_val+1 == n[i]:
        start_val = start_val+1
```
